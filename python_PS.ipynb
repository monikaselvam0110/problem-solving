```python
#remove all zeros from the given  number
n=input("Enter the number:")

for i in n:
  if i!='0':
    print(i,end="")
```

```python
#print numbers 1 to n
n=int(input())
for i in range(1,n+1):
  print(i,end=" ")
```

```python
#print even numbers upto N
n=int(input())
for i in range(1,n+1):
  if i%2==0:
    print(i,end=" ")
```

```python
#sum of n natural numbers
n=int(input())
sum=n*(n+1)//2
print(sum)
```

```python
#multiplication table
n=int(input())
for i in range(1,11):
  print(n,"X",i, "=",n*i)
```

```python
#factorial of a number
n=int(input())
fact=1
for i in range(1,n+1):
  fact *= i
print(fact)
```

```python
#count digits in given number
n=int(input())
count=0
while n>0:
  count+=1
  n=n//10

print(count)


```

```python
#reverse a number
n=int(input())
print(str(n)[::-1])
```

```python
#reverse number
n=int(input())
rev=0

while n >0 :
  digit=n%10 #takes last digit
  rev=rev*10 +digit
  n=n//10 #//10->removes last digit
print(rev)
```

```python
#sum of the digits
n=int(input("Enter the digits:"))
sum=0
while n >0:
  digit=n%10
  sum+=digit
  n=n//10

print(sum)
```

```python
#squares from  1 to N
n=int(input())

for i in range(1,n+1):
  print(i*i,end=" ")
```

```python
#odd numbers in a range
a=int(input())
b=int(input())
for i in range(a,b+1):
  if i%2!=0:
    print(i,end=" ")
```
