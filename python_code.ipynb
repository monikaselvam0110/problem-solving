***SWITCH***

```python
a = 10
b = 5
choice = '+'

match choice:
    case '+':
        print("Addition =", a + b)
    case '-':
        print("Subtraction =", a - b)
    case '*':
        print("Multiplication =", a * b)
    case '/':
        print("Division =", a / b)
    case _:
        print("Invalid Choice")
```

```python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
day = 6

if day >= 1 and day <= 5:
    print("Weekday")
elif day == 6 or day == 7:
    print("Weekend")
else:
    print("Invalid Day")
```

```python
month = 12

match month:
    case 12 | 1 | 2:
        print("Winter")
    case 3 | 4 | 5:
        print("Spring")
    case 6 | 7 | 8:
        print("Summer")
    case 9 | 10 | 11:
        print("Autumn")
    case _:
        print("Invalid Month")
```

```python
balance = 5000
choice = 1

match choice:
    case 1:
        print("Balance =", balance)

    case 2:
        withdraw = 1000
        if withdraw <= balance:
            balance -= withdraw
            print("Withdraw Successful")
            print("Balance =", balance)
        else:
            print("Insufficient Balance")

    case 3:
        deposit = 2000
        balance += deposit
        print("Deposit Successful")
        print("Balance =", balance)

    case _:
        print("Invalid Choice")
```
